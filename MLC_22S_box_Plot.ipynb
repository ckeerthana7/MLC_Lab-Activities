The rule:
This is called Tukeys Rule (used in box plots).


Used
   - Overiew of data
   - Identify Outliers
   - Distrubution of data
   - Spread of data

when no to use

    - less data
    - to know individual data
    - extreme values significantly different from other values
    - different scales
    - non-numeric values

```python
# Find outlier cereals in sugars using the IQR rule.
Q1 = df["sugars"].quantile(0.25)
Q3 = df["sugars"].quantile(0.75)
IQR = Q3 - Q1

low = Q1 - 1.5 * IQR
high = Q3 + 1.5 * IQR

outliers = df[(df["sugars"] < low) | (df["sugars"] > high)]
print(outliers[["name", "sugars"]] if "name" in df.columns else outliers[["sugars"]])
```
