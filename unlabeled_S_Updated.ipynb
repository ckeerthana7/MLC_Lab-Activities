# Anatomy of Dataframe

image.png

The three components of a DataFrame

    index
    columns
    data or values.

The index represents the sequence of values on the far left-hand side of the DataFrame.
Sometimes the index is referred to as the row labels.

The columns are the sequence of values at the very top of the DataFrame. Referred to as column name or column label.
