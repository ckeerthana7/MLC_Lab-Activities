```
import matplotlib.pyplot as plt
```

```
df['fueltype'].value_counts().plot(kind='bar')
plt.show()
```

```
df['carname'].value_counts()
```

```
df['company'] = df['carname'].str.split(' ').str[0]
stack=pd.crosstab(df['company'],df['carname'])
```

```
stack.plot(kind='bar', stacked = True, figsize=(14,7), legend=False)
plt.xlabel("Comapny")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
