Pie Chart


    - It is represented in circular graph splited in to smaller section.
    - Part of section is called as wedges or slices.(The segment represents each categories)
    - To show the relative size of data.


Application


    - used to show the relationships of parts to the whole variable.
    - It is used to represent only one series of data.


piechart.png



    When to use it ?

    - Show percentage or proportional data.
    - Less than 7 categories.
    - Display data that is classified into nominal or ordinal categories.
    - Try to use positive values.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Course': ['Electronics', 'Cloud', 'Cybersecurity', 'AIML'],
    'Students': [35, 25, 14, 15]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
plt.pie(df.Students, labels = df.Course, autopct='%1.2f%%')
#plt.pie(df.Students, labels = df.Course)
plt.show()
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

# Plot values using Pie Chart
# autopct enables you to display the percent value

plt.pie(sizes, labels = labels, autopct='%1.2f%%')


# To add Title
plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

# To change colors
plt.pie(sizes, labels = labels, autopct='%1.2f%%',colors=['violet','lightgreen','tomato','r'])

plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
# Explode: To emphasis the slices
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

plt.pie(sizes, labels = labels, autopct='%1.2f%%',colors=['violet','lightgreen','tomato','r'],explode=(0.2,0.1,0,0))

plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

# To change the figure size
plt.figure(figsize=(8,6))

# to change the start angle
plt.pie(sizes, labels = labels, autopct='%.2f%%',colors=['violet','lightgreen','tomato','m'],explode=(0.2,0,0,0), shadow=True, startangle=90)

plt.title('Breakdown of Student Strength by Course')

# To add Legend
plt.legend(labels, loc='lower right',title="course name:")
plt.show()
```

Variation - Donut Chart

          A donut chart is a variant of the pie chart, with a blank center allowing for additional information about the data as a whole to be included.

          This type of circular graph can support multiple statistics at once and it provides a better data intensity ratio to standard pie charts.

          It does not have to contain information in the center.


```python
# To create Circle
circle = plt.Circle((0, 0), 0.7, color='w')


d = plt.pie(sizes, labels=labels, autopct='%.1f%%',
            startangle=90, labeldistance=0.95)

# To add Circle in Pie chart
plt.gca().add_artist(circle)

plt.show()
```

```python
#Filter cereals with sugar content greater than 10g. Create a pie chart showing manufacturer distribution among these high-sugar cereals. Which manufacturer contributes most to this category?
plt.pie(df.sugars[df.sugars>10].value_counts(),labels=df.mfr[df.sugars>10].value_counts().index,autopct='%1.2f%%')
plt.show()
```

```python
#Create a pie chart to show the manufacturer distribution among healthy cereals.
plt.pie(df.mfr[df.healthy=='yes'].value_counts(),labels=df.mfr[df.healthy=='yes'].value_counts().index,autopct='%1.2f%%')
plt.show()
```
