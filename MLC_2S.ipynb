Descriptive Analytics (What? How?)

What is the average final mark of the students?

How many students have attendance above 80%?

What is the highest and lowest quiz average?

How many students scored above 75 in the final exam?

Diagnostic Analytics (Why?)

Why do students with attendance above 85% tend to have higher final marks?

Is there a relationship between assignment average and final marks?

Why did Student S2 score significantly lower compared to others?

Do quiz scores or assignment scores contribute more to final marks?
