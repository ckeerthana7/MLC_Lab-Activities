Ask Questions:

Are the packets increasing fast?

How many Ip are there?

What are the protocols used?

Is there any attack?

Process:

Average packets/sec:  586 packets/sec,
Increase from 125 to 893 avg after sec 2,
All TCP from single IP suggests targeted flood.

Information:

Packet rate from 10.0.0.5 increases rapidly to 920/sec on TCP, indicating anomalous traffic surge exceeding normal baselines.
​

Interpretation:

Sudden spike signals potential reconnaissance scan or volumetric attack initiation, as sustained high-rate TCP from one source deviates from legit flows.

Knowledge:

Pattern matches DDoS: single-source TCP floods often precede exploits; validate via baseline and threat intel.
​

Decision:

Block the IP
